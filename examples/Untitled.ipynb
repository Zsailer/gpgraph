```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from gpmap.simulate import MountFujiSimulation
import gpgraph as gpg
```

```python
sim = MountFujiSimulation.from_length(4)
sim.set_roughness((-1,1))
```

```python
G = gpg.GenotypePhenotypeGraph()
G.add_gpm(sim)

fig, ax = plt.subplots(figsize=(3,2))
x = gpg.draw_flattened(G, ax=ax, with_labels=False, node_size=100)
fig.savefig('readme-fig.png', format='png', dpi=250)
```
