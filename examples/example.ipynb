```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from gpmap import GenotypePhenotypeMap

from gpgraph import GenotypePhenotypeGraph
from gpgraph.draw import draw_paths
from gpgraph.paths import forward_path_prob
from gpgraph.models import strong_selection_weak_mutation
```

```python
gpm = GenotypePhenotypeMap.read_json('weinreich.json')
```

```python
G = GenotypePhenotypeGraph(gpm)

G.add_model(strong_selection_weak_mutation)
p = forward_path_prob(G, source="00000", target="11111")


fig, ax = plt.subplots(figsize=(5,5))
x = draw_paths(
    G, p, 
    ax=ax, 
    edge_scalar=9e6, 
)
#fig.savefig('readme-fig.png', format='png', dpi=250)
```
